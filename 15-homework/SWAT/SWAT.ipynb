## Step 2

Import the TXT files into Python, use RegEx to clean up the mess and get all the numbers. Put them into a list of dictionaries and eventually into a dataframe:

```python
import re
import pandas as pd
%matplotlib inline
```

```python
data=[]

for pdf in pdfs:
    with open('out' + pdf[2:-4] + '.txt', 'r') as myfile:
        content = myfile.read()
    data.append({
        'year': pdf[-8:-4],
        'high_risk': re.search('High Risk Warrant Service – (\d{1,3})', content).group(1),
        'tactical': re.search('Tactical Operations – (\d{1,3})', content).group(1),
        'special': re.search('Special Events – (\d{1,3})', content).group(1),
        'outside': re.search('Outside Agency Assists – (\d{1,3})', content).group(1)
    })

df = pd.DataFrame(data)

for c in df.columns:
    df[c] = pd.to_numeric(df[c])

df
```

## Step 3

Playing around with the data …

```python
# sums for each year
df['sum'] = df['high_risk'] + df['outside'] + df['special'] + df['tactical']
df
```

```python
# total sum 2000-2010
df['sum'].sum()
```

```python
# percentages of the different operation types
for c in df.columns[0:4]:
    print(c + ':', round(df[c].sum() / df['sum'].sum() * 100, 1), 'percent')
```

```python
# plot sum
df.plot(kind='bar', x='year', y='sum')
```
